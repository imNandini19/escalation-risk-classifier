In our dataset we are having nearly 1.3% high complaints and remaining all are low risk complaints. so if we train our model with this highly imbalanced data then our model will definitely give good accuracy but the model will fail for F1 score as our input data is highly imbalanced.

 To avoid our data being highly imbalanced I'm Redefining the label using Some escalation keywords like


*   fraud
*   complaint filed
*   legal
*   police
*   law suit
*   court

and some other more keywords



```python
# here is the exact list of escalation keywords
escalation_keywords = [
    "lawsuit",
    "legal",
    "court",
    "attorney",
    "lawyer",
    "fraud",
    "scam",
    "police",
    "complaint filed",
    "cfpb",
    "ftc",
    "harassment",
    "threat",
    "sue",
    "discrimination",
    "identity theft"
]
```

```python
# for each of the customer complaint first we will check whether his complaint is having any of the escalation keywords
# here is the function for searching keyword in compaint using re module
import re

def contains_escalation_keywords(text):
    text = text.lower()
    for word in escalation_keywords:
        if re.search(r"\b" + re.escape(word) + r"\b", text):
            return 1
    return 0
```

```python
df = df[df["Consumer complaint narrative"].notna()].copy()
df.shape
```

```python
df["keyword_flag"] = df["Consumer complaint narrative"].apply(contains_escalation_keywords)

df["timely_flag"] = df["Timely response?"].apply(lambda x: 1 if x == "No" else 0)

df["risk_label"] = df.apply(
    lambda row: 1 if row["keyword_flag"] == 1 or row["timely_flag"] == 1 else 0,
    axis=1
)
```

## Model 1: Training using TF-IDF + Logistic

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
```

```python
X=df['Consumer complaint narrative']
y=df['risk_label']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape , X_val.shape

# Since my dataset was imbalanced, I used stratified sampling to ensure both train and test sets maintain the same class distribution, preventing biased evaluation.
```

```python
#TF-IDF
vectorizer = TfidfVectorizer(
    max_features=10000,
    ngram_range=(1,2),
    stop_words="english"
)
```

```python
X_train_tfidf = vectorizer.fit_transform(X_train)
X_val_tfidf = vectorizer.transform(X_val)
```

```python
model = LogisticRegression(
    class_weight="balanced",
    max_iter=1000
)
```

```python
model.fit(X_train_tfidf, y_train)

y_pred = model.predict(X_val_tfidf)
```

```python
print(classification_report(y_val, y_pred))
```

Instead of using the default 0.5 decision threshold, I tuned the classification threshold using the precision-recall curve. Lowering the threshold to 0.41 improved high-risk recall from 0.76 to 0.87 while maintaining acceptable precision at 0.70.

```python
from sklearn.metrics import precision_recall_curve

y_probs = model.predict_proba(X_val_tfidf)[:, 1]

precisions, recalls, thresholds = precision_recall_curve(y_val, y_probs)
```

```python
best_threshold = None

for p, r, t in zip(precisions, recalls, thresholds):
    if r >= 0.80 and p >= 0.70:
        best_threshold = t
        print("Threshold:", t)
        print("Precision:", p)
        print("Recall:", r)
        break

print("Selected Threshold:", best_threshold)
```

## conclusion

#  Baseline Model: TF-IDF + Logistic Regression

Before jumping into transformers, I built a strong linear baseline to understand how well simple text features can detect escalation risk.

#  What This Model Does

Converts complaint text into TF-IDF vectors (unigrams + bigrams)

Uses Logistic Regression for classification

Handles class imbalance using class_weight="balanced"

The goal was to establish a realistic performance benchmark before trying more complex models like DistilBERT.

#  Dataset Summary

Total usable complaints: 5,485

High Risk: ~32%

Low Risk: ~67%

Average complaint length: ~1,047 characters

Labels were engineered using:

Untimely company responses

Escalation-related keywords (e.g., legal, fraud, lawsuit)

This is a weakly supervised setup, which reflects how labels are often constructed in real-world ML systems.

#  Performance with Default Threshold (0.5)
Class	Precision	Recall	F1-score
Low Risk (0)	0.89	0.93	0.91
High Risk (1)	0.83	0.76	0.79

Overall Accuracy: 87%

High-risk recall: 76%

This means the model correctly identifies most escalation cases, but still misses about 24% of them.

 Threshold Optimization

Instead of keeping the default threshold (0.5), I tuned it using the PrecisionRecall curve.

At a threshold of 0.41:

High-risk Precision: 0.70

High-risk Recall: 0.87

This significantly reduces missed escalation cases (from 24% down to 13%), at the cost of more false positives.

From a business perspective, catching more high-risk complaints is often more important than minimizing false alarms.

#  Model Behavior

The model learned strong escalation signals such as:

Legal terminology (e.g., attorney, lawsuit)

Fraud-related language

Identity theft references

One advantage of the linear model is interpretability  we can directly inspect which words contribute most to high-risk predictions.

#  Key Takeaways

A simple TF-IDF + Logistic Regression model performs surprisingly well.

Threshold tuning significantly improves escalation detection.

The model is lightweight, fast, and interpretable.

Any transformer-based model must clearly outperform this baseline to justify added complexity.
